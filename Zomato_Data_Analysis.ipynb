```python
 country_code = pd.read_excel('Country-Code.xlsx')
 country_code
```

Our csv file has 2 columns-
'*Country Code*' and 
'*Country*' 

 As you can recall, we have the '*Country Code*' column in our CSV file.
 
So let us combine this Country-Code file and our dataset(zomato.csv file) using the pandas **merge function **on this column

```python
#merge function to merge both CSV and XSL files

#left join focusing on the left (csv dataset)

final_df = pd.merge(zomato , country_code , on='Country Code',how='left')

final_df

   
```

TOP 3 Countries as per the count of Restaurants

```python
country_counts = final_df['Country'].value_counts()
top_3_country_counts = country_counts.index[:3]
top_3_country_counts 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.pie(country_counts[:3] , labels = top_3_country_counts  , autopct='%1.2f%%')
plt.plot()
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})

ratings
```

INFERENCE--> Around 2k ppl have given zero ratings.

 ratings (4.5-4.9) = excellent. 
 
 4.0 to 4.4 = v.good.
 
 3.5 to 3.9 = good.
  
 2.5 to 3.4 = average & so on

 

We conclude the following color assigned to the ratings:

Rating 0 - White - Not rated

Rating 1.8 to 2.4  -  Red - Poor

Rating 2.5 to 3.4  -  Orange - Average

Rating 3.5 to 3.9  -  Yellow - Good

Rating 4.0 to 4.4  -  Green - Very Good

Rating 4.5 to 4.9  -  Dark Green - Excellent

```python
import plotly.express as px
from plotly.offline import init_notebook_mode, plot, iplot

fig = px.bar(ratings,  x='Rating color', color='Rating text',y='Rating Count', color_discrete_sequence=['White', 'Red','Orange','Yellow','Green',"Darkgreen"], title='Rating Distribution')

fig.update_layout(bargap=0.1)

fig.show()
```

Observations: But clearly, among all the rated restaurants, orange (average) is most frequent

 Not rated count is very high, we need to see what might be the cause


```python
No_rating = final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index().rename( columns={0:'Rating Count'} )
descending_no_rating  = No_rating.sort_values('Rating Count' , ascending = False)
descending_no_rating
```

Observations:     

India seems to have maximum unrated restaurants. 

The reason might be that, in India...... the culture of ordering online food is still gaining momentum hence most of the restaurants are still unrated on Zomato....... as people might be yet not accoustomed to give ratings.
